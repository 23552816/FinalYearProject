```python
%%time
Sheet_credential = gspread.service_account("client_key.json")
 
# Open Spreadsheet by URL
# spreadsheet = Sheet_credential.open_by_url('paste your sheet url')
 
# Open Spreadsheet by key
spreadsheet = Sheet_credential.open_by_key('1leZ5C3hmBSnmRUutOR4uobGVPoskhxgptw_CEjAP8zo')
print(spreadsheet.title)

method = 0  # 0 = SVM ; 1 = ARIMA
hyperparam_set = 0

user_index=1
while os.path.exists('../data/user'+str(user_index-1)+"_"+str(hyperparam_set)+".csv"):
    user_index+=1   

real_store = np.zeros((user_index,48), dtype=float)
pred_store = np.zeros((user_index,48), dtype=float)

user_index=0
while os.path.exists('../data/user'+str(user_index)+"_"+str(hyperparam_set)+".csv"):
    
    energy = load_data('../data',user_index,hyperparam_set)[['load']]
    
    train_start_dt = '2022-12-25  00:30:00'
    test_start_dt = '2022-12-30  12:00:00'
 
    if method == 1: test_start_dt = '2022-12-31  00:30:00'
    
    #Hyperparameter Set
    if hyperparam_set == 1:
        train_start_dt = '2022-10-25  00:30:00'
        test_start_dt = '2022-10-30  12:00:00'

        if method == 1: test_start_dt = '2022-10-31  00:30:00'
    
    energy[(energy.index <= test_start_dt) & (energy.index >= train_start_dt)][['load']].rename(columns={'load':'train'}) \
    .join(energy[test_start_dt:][['load']].rename(columns={'load':'test'}), how='outer') \
    .plot(y=['train', 'test'], figsize=(15, 8), fontsize=12)
    plt.xlabel('timestamp', fontsize=12)
    plt.ylabel('load', fontsize=12)
    plt.show()
    
    train = energy.copy()[(energy.index >= train_start_dt) & (energy.index < test_start_dt)][['load']]
    test = energy.copy()[energy.index >= (test_start_dt)][['load']]
    
    scaler = MinMaxScaler()
    train['load'] = scaler.fit_transform(train)    
    test['load'] = scaler.transform(test)    
    if method == 1:
    
        HORIZON = 1

        test_shifted = test.copy()

        for t in range(1, HORIZON):
            test_shifted['load+'+str(t)] = test_shifted['load'].shift(-t, freq='30T')
    
        test_shifted = test_shifted.dropna(how='any')
        
        training_window = 144 # dedicate 6 days for training

        train_ts = train['load']
        test_ts = test_shifted

        history = [x for x in train_ts]
        history = history[(-training_window):]

        predictions = list()

        order = (4, 1, 0)
        seasonal_order = (1, 1, 0, 48)

        for t in range(test_ts.shape[0]):
            model = SARIMAX(endog=history, order=order, seasonal_order=seasonal_order)
            model_fit = model.fit()
            yhat = model_fit.forecast(steps = HORIZON)
            predictions.append(yhat)
            obs = list(test_ts.iloc[t])
            # move the training window
            history.append(obs[0])
            history.pop(0)
        
        eval_df = pd.DataFrame(predictions, columns=['t+'+str(t) for t in range(1, HORIZON+1)])
        eval_df['timestamp'] = test.index[0:len(test.index)-HORIZON+1]
        eval_df = pd.melt(eval_df, id_vars='timestamp', value_name='prediction', var_name='h')
        eval_df['actual'] = np.array(np.transpose(test_ts)).ravel()
        eval_df[['prediction', 'actual']] = scaler.inverse_transform(eval_df[['prediction', 'actual']])
        
        

        
        y_test = eval_df[eval_df['h'] == 't+1']['actual']
        y_test_pred = eval_df[eval_df['h'] == 't+1']['prediction']
        test_timestamps = test.index[0:len(test.index)-HORIZON+1]


        
        ## Plotting single step forecast
        eval_df.plot(x='timestamp', y=['actual', 'prediction'], style=['r', 'b'], figsize=(15, 8))
    
        plt.xlabel('timestamp', fontsize=12)
        plt.ylabel('load', fontsize=12)
        plt.show()
        
        
    elif method == 0:             
        

        train_data = train.values
        test_data = test.values

        timesteps=26

        train_data_timesteps=np.array([[j for j in train_data[i:i+timesteps]] for i in range(0,len(train_data)-timesteps+1)])[:,:,0]

        test_data_timesteps=np.array([[j for j in test_data[i:i+timesteps]] for i in range(0,len(test_data)-timesteps+1)])[:,:,0]

        x_train, y_train = train_data_timesteps[:,:timesteps-1],train_data_timesteps[:,[timesteps-1]]
        x_test, y_test = test_data_timesteps[:,:timesteps-1],test_data_timesteps[:,[timesteps-1]]

        model = SVR(kernel='rbf',gamma=0.1, C=5, epsilon = 0.05) #SVR(kernel='rbf',gamma=0.5, C=10, epsilon = 0.05)
        model.fit(x_train, y_train[:,0])

        y_train_pred = model.predict(x_train).reshape(-1,1)
        y_test_pred = model.predict(x_test).reshape(-1,1)

        y_train_pred = scaler.inverse_transform(y_train_pred)
        y_test_pred = scaler.inverse_transform(y_test_pred)

        y_train = scaler.inverse_transform(y_train)
        y_test = scaler.inverse_transform(y_test)


        train_timestamps = energy[(energy.index < test_start_dt) & (energy.index >= train_start_dt)].index[timesteps-1:]
        test_timestamps = energy[test_start_dt:].index[timesteps-1:]
        
        plt.figure(figsize=(25,6))
        plt.plot(train_timestamps, y_train, color = 'red', linewidth=2.0, alpha = 0.6)
        plt.plot(train_timestamps, y_train_pred, color = 'blue', linewidth=0.8)
        plt.legend(['Actual','Predicted'])
        plt.xlabel('Timestamp')
        plt.title("Training data prediction")
        plt.show()
        print('MAPE for training data: ', round(mape(y_train_pred, y_train)*100, 2), '%')
        

    row_vector_test = np.array(y_test)
    row_vector_test = np.transpose(row_vector_test) 
    real_store[user_index+1,:] = row_vector_test
    real_store[0] = real_store[0] + row_vector_test 
    
    
    plt.figure(figsize=(10,3))
    plt.plot(test_timestamps, y_test, color = 'red', linewidth=2.0, alpha = 0.6)
    plt.plot(test_timestamps, y_test_pred, color = 'blue', linewidth=0.8)
    plt.legend(['Actual','Predicted'])
    plt.xlabel('Timestamp')
    plt.show()    
    print('MAPE for testing data: ', round(mape(y_test_pred, y_test)*100, 2), '%')
    
    
    y_test_pred_row_vector = y_test_pred.T
    pred_store[user_index+1,:] = y_test_pred_row_vector
    pred_store[0] = pred_store[0] + y_test_pred_row_vector
    user_index+=1   
```

```python
def calculate_metrics(a, b, j):
    # Ensure that both arrays have the same shape
    if a.shape != b.shape:
        raise ValueError("Arrays 'a' and 'b' must have the same shape")
        
    mask_a = (a > j)
    mask_b = (b > j)

    count_a = np.sum(mask_a)
    count_b = np.sum(mask_b)
    
    error_mask = (mask_a != mask_b)
    error_count = np.sum(error_mask)

    # Create boolean masks for true positive, false positive, true negative, and false negative
    true_positive = np.sum((a > j) & (b > j))
    false_positive = np.sum((a <= j) & (b > j))
    true_negative = np.sum((a <= j) & (b <= j))
    false_negative = np.sum((a > j) & (b <= j))

    # Calculate Sensitivity (True Positive Rate)
    sensitivity = true_positive / (true_positive + false_negative)

    # Calculate Specificity (True Negative Rate)
    specificity = true_negative / (true_negative + false_positive)

    # Calculate Precision (Positive Predictive Value)
    precision = true_positive / (true_positive + false_positive)

    return count_a, count_b, error_count, sensitivity, specificity, precision



max_usage_threshold = 28000/38196*np.max(real_store[0]) #70 with only 4 normalized


arr_threshold = np.full((1, 48), max_usage_threshold)
plt.figure(figsize=(10,3))
plt.plot(test_timestamps, arr_threshold[0], color = 'green', linewidth=2.0, alpha = 0.4)
plt.plot(test_timestamps, real_store[0], color = 'red', linewidth=2.0, alpha = 0.6)
plt.plot(test_timestamps, pred_store[0], color = 'blue', linewidth=0.8)
plt.legend(['Maximum Power Usage Threshold','Actual','Predicted'])
plt.xlabel('Timestamp')
plt.ylabel('Load (kVA)')
plt.show()
print('MAPE for total testing data: ',round(mape(pred_store[0], real_store[0])*100, 2) , '%')

count_pred, count_real, error_count, sensitivity, specificity, precision = calculate_metrics(pred_store[0], real_store[0], max_usage_threshold)
print("Calculated NMD:", round(28000/38196*np.max(real_store[0]), 2))
print("Number of exceedances in prediction:", count_pred)
print("Number of exceedances in real:", count_real)
print("Number of errors:", error_count)
print("Sensitivity/Recall:", round(sensitivity, 2))
print("Specificity:", round(specificity, 2))
print("Precision:", round(precision, 2))

print("Average Real Load:", round(np.mean(real_store[0]), 2))
print("Max Real Load:", round(np.max(real_store[0]), 2))

arr_exceed = np.where(pred_store[0] > max_usage_threshold)[0]
timestamps_new = test_timestamps.strftime('%Y-%m-%d %H:%M:%S')
strings_exceeding_thresh = [timestamps_new[i] for i in arr_exceed]


row_vector_indices = np.zeros((len(strings_exceeding_thresh),3), dtype=float)
rvii = 0


for exceedance_index in arr_exceed:
    timestamp_values = pred_store[1:, exceedance_index]
   # print(timestamp_values)
    sorted_indices = np.argsort(timestamp_values)[::-1] 
  #  print(sorted_indices)
    top_three_indices = sorted_indices[:3] 
   # print(top_three_indices)
    row_vector_indices[rvii] = np.array(top_three_indices) 
   # print("Indices of the Three Highest Values:", row_vector_indices)
    rvii+=1
    
#print(arr_exceed)
#print(test_timestamps)
```

```python
import ipywidgets as widgets
import matplotlib.pyplot as plt
%matplotlib inline

# Create a list of time options in 30-minute intervals from 00:30 to 24:00
time_options = []
for hour in range(24):
    for minute in range(0, 60, 30):
        if hour == 0 and minute == 0:
            continue  # Skip 00:00

        time_options.append(f"{hour:02}:{minute:02}")
time_options.append('24:00')



# Create a Dropdown widget with time options
time_dropdown = widgets.Dropdown(
    options=time_options,
    description='Select Time:',
)

user_labels = [x for x in range(0, user_index)]

# Create a function to handle the widget's value change event
def on_time_change(change):
    
    selected_time = change.new
    index = time_options.index(selected_time)
    catagories = [f"user{index}" for index in range(0, user_index)]
    values = np.zeros(user_index, dtype=float)
    x_pos = list(range(0, user_index))

    selected_column = pred_store[1:user_index+1, index:index+1].T
    values = selected_column[0, :].flatten()
    values = values.tolist()

    values = [(value / sum(values)) * 100 for value in values]
   # clear_output(wait=True)

    plt.bar(x_pos, values, alpha=0.5)

    plt.title('Total predicted usage per user at time: '+ selected_time) 
    plt.xlabel('User') 
    plt.ylabel('% Contribution to Total Load') 
    plt.xticks(user_labels, catagories) 
    plt.grid(axis='y')

    plt.show()
    

# Attach the event handler to the widget
time_dropdown.observe(on_time_change, names='value')

# Display the widget
time_dropdown
```
